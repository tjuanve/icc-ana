```python
import pandas as pd
import numpy as np
```

```python
file_path = "/data/user/rnaab/GlobalFit/NNMFit/datasets/SnowStorm_example_tracks/dataset_example.hdf"

df = pd.read_hdf(file_path)
```

```python

# see https://link.springer.com/article/10.1140/epjc/s10052-024-13137-2/figures/7
effective_area_m = 4.0e3 # arca230 horizontal numu at 50 PeV [m^2] 

effective_area_cm = effective_area_m*1e4

df.iloc[1].keys()

n_events = 5

df_toy = df.iloc[:n_events].copy()

i=0
for index, row in df_toy.iterrows():
    df_toy.at[index, 'energy_truncated'] = 4.5e7 + i*0.2e7
    df_toy.at[index, 'MCPrimaryEnergy'] = 4.5e7 + i*0.2e7
    df_toy.at[index, 'zenith_MPEFit'] = np.pi/2 - 5*np.pi/180 + i*5*np.pi/180/n_events
    df_toy.at[index, 'MCPrimaryZenith'] = np.pi/2 - 5*np.pi/180 + i*5*np.pi/180/n_events

    df_toy["OneWeight"] = effective_area_cm/n_events
    df_toy["powerlaw"] = np.nan
    df_toy["fluxless_weight"] = np.nan

    i+=1
    
# df_toy["energy_truncated"] = 2e7
# df_toy["MCPrimaryEnergy"]  = 2e7
# df_toy["zenith_MPEFit"]  = np.pi/2 # horizontal
# df_toy["MCPrimaryZenith"]  = np.pi/2


# df_toy["OneWeight"] = effective_area_cm
# df_toy["powerlaw"] = np.nan
# df_toy["fluxless_weight"] = np.nan

df_toy
```
