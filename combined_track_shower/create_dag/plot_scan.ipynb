```python
scan = NNMFit.utilities.ScanHandler(indir="/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/create_dag/output")
scan_track = NNMFit.utilities.ScanHandler(indir="/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/create_dag/output_track")
scan_cscd = NNMFit.utilities.ScanHandler(indir="/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/create_dag/output_cscd")
scan_track_cscd = NNMFit.utilities.ScanHandler(indir="/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/create_dag/output_track_cscd")
```

```python
drop_unsuccessful = False

x = {}
y = {}
dllh = {}

x["bench"], y["bench"], dllh["bench"] = scan.get_scan_xydllh(x="astro_norm",y="gamma_astro",interpolate=True, drop_unsuccessful=drop_unsuccessful)
x["track"], y["track"], dllh["track"]  = scan_track.get_scan_xydllh(x="astro_norm",y="gamma_astro",interpolate=True, drop_unsuccessful=drop_unsuccessful)
x["cscd"], y["cscd"], dllh["cscd"]  = scan_cscd.get_scan_xydllh(x="astro_norm",y="gamma_astro",interpolate=True, drop_unsuccessful=drop_unsuccessful)
x["track_cscd"], y["track_cscd"], dllh["track_cscd"] = scan_track_cscd.get_scan_xydllh(x="astro_norm",y="gamma_astro",interpolate=True, drop_unsuccessful=drop_unsuccessful)

```

```python
import matplotlib.pyplot as plt

# levels = [2.3, 6.18]  # 1sigma (68%), 2sigma (95%)
levels = [2.3]  # 1sigma (68%), 2sigma (95%)

fig, (ax) = plt.subplots(1, 1)

# pcm = ax.pcolor(x["bench"], y["bench"], dllh["bench"], cmap='viridis', shading='auto')
# pcm = ax.pcolor(x["track"], y["track"], dllh["track"], cmap='viridis', shading='auto')

# fig.colorbar(pcm, ax=ax,label=r"dllh")

cntr = ax.contour( x["bench"], y["bench"], dllh["bench"], levels=levels, colors=["black"], linewidths=3.5, label = "bench" )
cntr_track = ax.contour( x["track"], y["track"], dllh["track"], levels=levels, colors=["red"], linewidths=3.5, label = "track" )
cntr_cscd = ax.contour( x["cscd"], y["cscd"], dllh["cscd"], levels=levels, colors=["green"], linewidths=3.5, label = "cscd" )
cntr_track_cscd = ax.contour( x["track_cscd"], y["track_cscd"], dllh["track_cscd"], levels=levels, colors=["blue"], linewidths=3.5, label = "all" )

ax.plot([1.44], [2.37],"*")

ax.set_xlabel(r"$\Phi_0$")
ax.set_ylabel(r"$\gamma$")

# Create custom legend
legend_elements = [
    plt.Line2D([0], [0], color="black", lw=3.5, label="bench"),
    plt.Line2D([0], [0], color="red", lw=3.5, label="track"),
    plt.Line2D([0], [0], color="green", lw=3.5, label="cscd"),
    plt.Line2D([0], [0], color="blue", lw=3.5, label="all"),
]

# Add legend to the plot
ax.legend(handles=legend_elements, loc="upper right")

plt.show()


```
