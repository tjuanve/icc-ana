```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import LogLocator
```

```python
import NNMFit
from NNMFit.utilities.readout_graphs import HistogramGraph
from NNMFit.core.analysis_config import AnalysisConfig
```

```python
det_config = "IC86_pass2_SnowStorm_v2_tracks"

config_hdl_dict = {}
energy_ranges = ["1_10PeV", "3_30PeV", "10_100PeV", "30_300PeV", "100_1000PeV"]

hist_graph_dict = {}

inputs = {}

res = {}

hist = {}

yerror = {}

bins_energy = {}
bins_zenith = {}



for energy_range in energy_ranges:
    config_hdl_dict[energy_range] = AnalysisConfig.from_configs(
        main_config_file=f"/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/GlobalFit_configs_extraSPL/main_brokenPL_extraSPL_{energy_range}.cfg",
        analysis_config_file = "/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/GlobalFit_configs_extraSPL/analysis_configs/data/data_SAY_he_nominal_emptyICC_10_100PeV.yaml",
        config_dir="/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/GlobalFit_configs_extraSPL",
        override_config_files=["override/cscd/Cscd_fullrange_drop1bin_cscd_signal.cfg", "override/tracks/Tracks_fullrange_drop5bins_5sys.cfg", "override/minuit.cfg", "override/cscd/Cscd_FullSample_baseline.cfg", "override/event_overlap/Tracks_NoCscd_baseline.cfg"],
        override_parameters_files=["override/parameters/piecewise_combined.yaml", "override/parameters/BPL_breakbound.yaml", "override/parameters/two_component.yaml"],
        override_components_files=["override/components/fixed_delta_gamma_reference_dropped_bins.yaml", "override/components/piecewise_combined.yaml", "override/components/components_Tracks_Fullrange_drop_5lowEbins.yaml"]
        )
    
    hist_graph_dict[energy_range] = HistogramGraph(config_hdl_dict[energy_range])

    inputs[energy_range] = {
    # 'astro_norm': 1.77,
    'astro_norm': 0.0,
    'gamma_1': 1.31,
    'gamma_2': 1.31,
    'e_break': 4.39,
    f'astro_he_norm_{energy_range}': 1,
    f'gamma_astro_he_norm_{energy_range}': 1,
    'conv_norm' : 0,
    'prompt_norm' : 0,
    'muon_norm' : 0,
    'muongun_norm' : 0,
    } 

    res[energy_range] = hist_graph_dict[energy_range].get_evaled_histogram(input_variables=inputs[energy_range], det_config=det_config, reshape=True)
    hist[energy_range] = res[energy_range]["mu"].sum(axis=1)
    yerror[energy_range] = np.sqrt(res[energy_range]["ssq"].sum(axis=1))

    print(energy_range, sum(hist[energy_range]))

    temp = hist_graph_dict[energy_range].get_binning(det_config=det_config)
    bins_energy[energy_range] = temp["reco_energy"]
    bins_zenith[energy_range] = temp["reco_zenith"]
```

```python
# Apply the custom style
plt.style.use('/data/user/tvaneede/plotting_style/km3net.mplstyle')
```

```python
# 1d energy icecube
fig, (ax1) = plt.subplots(1, 1)

# energy: axis=1

colors = ["C0","C3","C2","C5","C1"]

for i,energy_range in enumerate(energy_ranges):
    ax1.stairs(values=hist[energy_range], edges=bins_energy[energy_range], color=colors[i],label=energy_range)
    ax1.stairs(values=hist[energy_range]+yerror[energy_range], edges=bins_energy[energy_range], baseline=hist[energy_range]-yerror[energy_range], alpha=0.5, color=colors[i])

# # the he event
# ax1.plot([6e7,6e7],[0, 2*max(hist)], color="black")
# ax1.plot([11e7,11e7],[0, 2*max(hist)], color="black")
# ax1.plot([32e7,32e7],[0, 2*max(hist)], color="black")

# some text
# plt.text(3.5e8, 2.5e-4, f"{det_config}\nlivetime: 0.1 yr\nEvents: {sum(hist):.5f}",horizontalalignment='right',)

# setup figure
ax1.set_xlabel(r"$E_{\rm reco}$ [GeV]")
ax1.set_ylabel(r"$N_{\mathrm{Events}}$")
ax1.set_xscale("log")
ax1.set_yscale("log")
# ax1.set_xlim(bins_energy[0], bins_energy[-1])

ax1.legend()

# plt.savefig(f"/data/user/tvaneede/GlobalFit/analysis/combined_track_shower/plots/output/ereco_{det_config}_norm{input_params['astro_norm']}_gamma{input_params['gamma_astro']}.pdf")

fig.tight_layout()
plt.show()
```
